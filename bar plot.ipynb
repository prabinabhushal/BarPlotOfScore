```python
import numpy as np
import matplotlib.pyplot as plt
#bar plot

x=["35-45 age","50-60 age"]
x=np.arange(len(x))
a=[60,75]  #height in kg
b=[141,35]   #height in cm
```

```python
plt.style.use("fivethirtyeight")
fig, ax = plt.subplots()

ax.bar(x,a,width=0.2) #blue
ax.bar(x+0.2,b,width=0.2)  #orange
```

```python
import pandas as pd
data = {
    "Teams":["A","B","C"],
    "Score1": [13,45,44],
    "Score2": [34,55,66],
    "Score3": [33,67,78]
}

df=pd.DataFrame(data)
df.plot(x="Teams",kind="bar",stacked=False)
```

```python
import pandas as pd
data = {
    "Teams":["A","B","C"],
    "Score1": [13,45,44],
    "Score2": [34,55,66],
    "Score3": [33,67,78]
}

df=pd.DataFrame(data)
df.plot(x="Teams",y= "Score1",kind="pie",stacked=True)
```

```python
plt.style.use("fivethirtyeight")
fig, ax = plt.subplots()
ax.pie([30,40,50,10],labels=["NewYork","Texas","Washington","Kansas"],explode=[0,0.2,0,0]);
```

```python
#pyplot

fig,ax = plt.subplots(2,figsize=(8,8)) #2 inch:width,2 inch:height
ax[0].plot([1,2,3,4])
ax[1].pie([30,40,20,10],labels=["NewYork","Texas","Washington","Kansas"],explode=[0,0.2,0,0]);
# ax.set_title()
# ax.set_xlabel()
# ax.set_ylabel()

# ax.set_xlim()
# ax.set_ylim()

fig.suptitle("This is pyplot")


# fig.savefig("example.png")
```
